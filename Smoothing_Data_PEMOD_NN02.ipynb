# Import Library


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
import pandas as pd
```

**Import Data**

```python
df = pd.read_excel('Data_Used.xlsx', header=None) # Import data dari excel ke dataset di pandas
df.columns = ['nilai']  # Beri nama kolom
data = df['nilai'].values # Inisiasi data
data = data[71:130] # Pilih range data yang diambil
```

**Smoothing With Savitzky-Golay Filter**

```python
df1 = pd.DataFrame(dict(x=data))
x_filtered = df1[["x"]].apply(savgol_filter,  window_length=31, polyorder=2)
plt.plot(data, label="Data Asli")
plt.plot(x_filtered, label="Data Smoothing dengan Savitzky-Golay Filter")
plt.title("Plot Smoothing Data dengan Savitzky-Golay Filter")
plt.xlabel("Waktu")
plt.ylabel("Orang Yang Terinfeksi")
plt.grid(True)
plt.legend()
plt.show()
```

Smoothing With Moving Averages

```python
data1 = pd.DataFrame(data, columns=['Value'])
moving_avg5 = data1['Value'].rolling(window=5).mean()
moving_avg10 = data1['Value'].rolling(window=10).mean()


plt.title("Plot Smoothing Data dengan Moving Average")
plt.xlabel("Waktu")
plt.ylabel("Orang Yang Terinfeksi")
plt.grid(True)
plt.plot(data1['Value'], label='Data Asli')
plt.plot(moving_avg5, label='Data Smoothing Moving Average dengan window 5')
plt.plot(moving_avg10, label='Data Smoothing Moving Average dengan window 10')
plt.legend()
plt.show()
```

```python
plt.title("Plot Smoothing Perbandingan Antara Savitzky-Golay Filter dan Moving Average")
plt.plot(x_filtered, label="Data Smoothing dengan Savitzky-Golay Filter")
plt.plot(moving_avg5, label='Data Smoothing Moving Average dengan window 5')
plt.xlabel("Waktu")
plt.ylabel("Orang Yang Terinfeksi")
plt.grid(True)
plt.legend()
plt.show()
```
